```python
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from PIL import Image
```

```python
OVERPASS_URL = 'http://overpass-api.de/api/interpreter'
```

To get all nodes that lay on a way in a radius of 200 meters around a point

```python
lat = 50.85866
lon = 3.29640
```

```python
# first get the ways around the point 

# list of dicts: node id with count of nodes
nodes = []
intersection_nodes = []

overpass_query = f"""[out:json];
                way(around: 100, {lat},{lon});
                out;"""

response = requests.get(OVERPASS_URL, params={'data': overpass_query})


for item in response.json()['elements']:
    print(item)
    for node in item['nodes']:
        if node not in nodes:
            nodes.append(node)
        else:
            if node not in intersection_nodes:
                intersection_nodes.append(node)

for node in intersection_nodes:
    print(node)



# node id 769368176 should be an intersection
```
