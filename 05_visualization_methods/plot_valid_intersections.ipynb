```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure for displaying the img_patches
fig = plt.figure(figsize=(10, 10))

# Iterate through validIntersection and display img_patch arrays
for i, intersection in enumerate(validIntersection):
    img_patch = [[intersection[1], intersection[2], intersection[3]],
                [intersection[0], 1, intersection[4]],
                [intersection[7], intersection[6], intersection[5]]]
    img_patch = np.array(img_patch) * 255

    # Create a new subplot for each img_patch
    ax = fig.add_subplot(len(validIntersection)//5 + 1, 5, i + 1)
    ax.imshow(img_patch, cmap='gray')
    ax.axis('off')  # Turn off axis

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the background and the padding between images
background_size = (len(validIntersection) // 11 + 1, 10)
padding = 3

# Create a gray background
full_image = np.full((background_size[0] * 3 + (background_size[0] - 1) * padding,
                      background_size[1] * 3 + (background_size[1] - 1) * padding),
                     128, dtype=int)

# Iterate through validIntersection and place img_patch arrays on the background
for i, intersection in enumerate(validIntersection):
    img_patch = [[intersection[1], intersection[2], intersection[3]],
                [intersection[0], 1, intersection[4]],
                [intersection[7], intersection[6], intersection[5]]]
    img_patch = np.array(img_patch) * 255

    row, col = divmod(i, background_size[1])
    y_start = row * (3 + padding)
    x_start = col * (3 + padding)

    full_image[y_start:y_start + 3, x_start:x_start + 3] = img_patch

# Display the full_image
plt.imshow(full_image, cmap='gray')
# plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.axis('off')  # Turn off the axis
plt.show()
plt.savefig("valid_intersections.png")
```
